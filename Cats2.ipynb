```python
X_train = df_train_new[['type_wild', 'group_group A',
       'group_group B', 'group_group C', 'group_group D', 'group_group E',
       "education_bachelor's degree", 'education_high school',
       "education_master's degree", 'education_some college',
       'education_some high school', 'meal_standard',
       'preparation course_none']]
y_train = df_train_new['Pass']
```

```python
X_test = df_test_new[['type_wild', 'group_group A',
       'group_group B', 'group_group C', 'group_group D', 'group_group E',
       "education_bachelor's degree", 'education_high school',
       "education_master's degree", 'education_some college',
       'education_some high school', 'meal_standard',
       'preparation course_none']]
```

```python
# 4b
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth=5, criterion='entropy')
clf = cross_val_score(model, X=X_train, y=y_train, cv=3, scoring='roc_auc')
```

```python
print(clf)
```

```python
fit_columns = ["cat_bio_('domestic', 'group A')", "cat_bio_('domestic', 'group B')",
       "cat_bio_('domestic', 'group C')", "cat_bio_('domestic', 'group D')",
       "cat_bio_('domestic', 'group E')", "cat_bio_('wild', 'a')",
       "cat_bio_('wild', 'group A')", "cat_bio_('wild', 'group B')",
       "cat_bio_('wild', 'group C')", "cat_bio_('wild', 'group D')",
       "cat_bio_('wild', 'group E')", "education_bachelor's degree",
       'education_high school', "education_master's degree",
       'education_some college', 'education_some high school', 'meal_standard',
       'preparation course_none']
```

```python
pred_proba = model.predict_proba(df_train_new[fit_columns])
```
