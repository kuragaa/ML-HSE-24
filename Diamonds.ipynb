# Работа с пропущенными значениями

Большинство моделей не умеют работать с пропусками, поэтому, если они есть в датасете - их нужно убрать или скорректировать

Стратегии заполнения пропусков:

* если пропусков очень мало (~меньше процента) - строки с пропусками можно удалить или заполнить средним/медианой

* если пропусков мало (~5-10%) - их можно заполнить (средним, медианой, уникальным значением, самым популярным значением, спрогнозировать)

* если пропусков много - можно удалить столбец

* можно пытаться предсказывать пропуски моделью

```python
df.isnull().mean() * 100 # среднее количество пропусков по столбцам (* на 100 - чтоб в процентах было)
```

В данном датасете пропусков < 1-2%

Пропущенных значений тут мало, поэтому в принципе их можно просто выкинуть. Если бы их было больше, то нужно было бы заменять.

Если мы про данные ничего не знаем, то лучше заменять пропуски медианным значением, а не средним, так как среднее значение не устойчиво к выбросам. То есть, если в датасете есть аномально большие или маленькие значния, то они очень сильно повлияют на среднее. А медиана - это просто некоторое центральное число в данных (центр в упорядоченной последовательности)

```python
df.hist(bins=100, figsize=(8,8)) # гистограмма по всем нашим числовым признакам (bins - число столбиков - чем больше, тем подробнее гистограмма)
```

По гистограмме можно приблизительно определить, есть ли у нас выбросы или какие-то хвосты. Но способ не самый надежный.

```python
median_carat = df['carat'].median() # считаем медианное значения по столбцу
#print(median_carat)
```

```python
df['carat'].fillna(median_carat, inplace=True) # заменяем все пропуски медианным значением в исходной таблице (в столбце караты)
```

```python
median_depth = df['depth'].median() # то же самое делаем со столбцом глубины

df['depth'].fillna(median_depth, inplace=True)
```

### Столбец - Число карат

* мелкие — массой до 0,30 карат (не включительно);
* средние — массой от 0,30 до 1 карат (не включительно);
* крупные – 1 карат и выше, называют солитеры.


Небольшие бриллианты различного размера, полученные при огранке алмаза, и весящие каждый в отдельности менее 0,25 карат, называют *melee (мили)*. Полученные таким же путем более крупные бриллианты различного размера называют *melange (меланж)*.


Бриллианты более 6 карат обычно продаются c аукционов.


Алмазам и бриллиантам более 25 карат присваивают собственные имена.

#### Группировка

Посчитаем среднюю стоимость алмазов в каждой категории

```python
df.groupby('cut')['price'].mean() # () - колонка, по которой делаем группировку. [] - на что смотрим в этой колонке
# замечаем прикол - в данном датасете алмазы плохой огранки fair самые дорогие в среднем. Это потому, что они больше весят. (факт - при огранке вес уходит)
```

Посчитаем несколько характеристик в зависимости от огранки

```python
df.groupby('cut').agg(PriceMean=('price','mean'),
                      PriceMedian=('price', 'median'),
                      CaratMean=('carat', 'mean'))
```

```python
df.groupby('cut').agg(PriceMean=('price','mean'),
                      PriceMedian=('price', 'median'),
                      CaratMean=('carat', 'mean')).sort_values(by='PriceMean', ascending=False)
# сортируем по PriceMean в порядке убывания (в итоге сортировка по убыванию средней стоимости)
```
