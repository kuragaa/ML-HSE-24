```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```python
df_train.head(10)
```

```python
df_train.info()
```

```python
# В скольких столбцах таблицы среднее значение измеряется в сотнях? (не в десятках и не в тысячах).
for col in df_train.columns:
    if df_train[col].dtype == object:
        continue
    print(df_train[col].mean())
```

```python
df_train['is_internet'] = 0
```

```python
# Если у телефона нет ни опции 3G, ни опции 4G, ни Wifi - то у пользователей нет доступа к интернету. Создайте колонку is_internet и поставьте туда 0, если у телефона нет ни одной из трех перечисленных опций, и 1 иначе.


df_train.loc[(df_train['wifi'] == 1) | (df_train['four_g'] == 1) | (df_train['three_g'] == 1), 'is_internet'] = 1
```

```python
df_train.head(10)
```

```python
#Такую же колонку создайте в тестовых данных.
df_test['is_internet'] = 0
df_test.loc[(df_test['wifi'] == 1) | (df_test['four_g'] == 1) | (df_test['three_g'] == 1), 'is_internet'] = 1
```

```python
# Какая доля телефонов (из train.csv) не имеет доступа к интернету? Ответ округлите до сотых.
 (len(df_train['is_internet']) - df_train['is_internet'].sum()) / len(df_train['is_internet'])
```

```python
# Какой столбец имеет наибольшую по модулю корреляцию Пирсона с целевой переменной price_range? В ответ запишите коэффициент корреляции (со знаком), округленный до сотых.
from sklearn.feature_selection import r_regression

coef = r_regression(df_train.drop(columns=['price_range', 'touch_screen']), df_train['price_range'])

np.max(np.abs(coef))
```

```python
# Среди телефонов ценовой категории 2 (price_range = 2) вычислите долю тех, которые не имеют bluetooth. Ответ округлите до сотых.
df_pr_2 = df_train[df_train['price_range'] == 2]
```

```python
df_pr_2_without_bluetooth = df_pr_2[df_pr_2['blue'] == 0]
```

```python
len(df_pr_2_without_bluetooth) / len(df_pr_2)
```

```python
# Среди телефонов с talk_time < 10 насколько больше телефонов с опцией dual_sim = 1, чем с dual_sim = 0?
df_talk = df_train[df_train['talk_time'] < 10]
df_talk_dual_sim_1 = df_talk[df_talk['dual_sim'] == 1]
df_talk_dual_sim_0 = df_talk[df_talk['dual_sim'] == 0]
np.abs(len(df_talk_dual_sim_1) - len(df_talk_dual_sim_0))
```

```python
# Разбейте тренировочные данные на целевой вектор y, содержащий значения из столбца price_range, и матрицу объект-признак X, содержащую остальные признаки.

Сколько изначально различных значений было в столбце touch_screen?
X_train = df_train.drop(columns='price_range')
y_train = df_train['price_range']
```

```python
X_test = df_test
```

```python
# #Посмотрите, сколько значений содержит категориальный столбец touch_screen.
X_test.touch_screen.value_counts()
```

```python
X_train.touch_screen.value_counts()
```

```python
# Значения, встречающиеся меньше, чем в 1% строк, замените на самое частое значение.
X_train['touch_screen'] = X_train['touch_screen'].apply(lambda x: 'Yes' if x == '-' else x)
```

```python
X_test['touch_screen'] = X_test['touch_screen'].apply(lambda x: 'No' if x == '-' else x)
```

```python
X_train.touch_screen.value_counts()
```

```python
X_test.touch_screen.value_counts()
```

```python
# После этого закодируйте столбец при помощи OneHot-encoding (используйте аргумент drop='first'). Обучайте OHE на тренировочных данных X, но кодируйте при этом и тренировочные, и тестовые данные.
from sklearn.preprocessing import OneHotEncoder

df_onehot = df_train.copy()

ohe = OneHotEncoder(drop='first', sparse_output = False)

encoded_df = pd.DataFrame(ohe.fit_transform(df_onehot[['touch_screen']]))
encoded_df
```

```python
encoded_df.info()
```

```python
print(list(encoded_df.columns))
```

```python
X_train['touch_screen'] = encoded_df[1]
```

```python
X_train.head()
```

```python
encoded_df = pd.DataFrame(ohe.transform(df_test[['touch_screen']]))
encoded_df
```

```python
X_test['touch_screen'] = encoded_df[1]
```

```python
# Для кросс-валидации лучше использовать функцию cross_val_score из библиотеки sklearn.model_selection.
# бучите на этих данных логистическую регрессию из sklearn (LogisticRegression) с параметрами по умолчанию. Выведите среднее значение метрики f1_score c вариантом усреднения ‘weighted’
# (или же 'f1_weighted') алгоритма на кросс-валидации с тремя фолдами. Ответ округлите до сотых.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

f1_scores = cross_val_score(LogisticRegression(
    penalty='l2',
    dual=False,
    tol=0.0001,
    C=1.0,
    fit_intercept=True,
    intercept_scaling=1,
    class_weight=None,
    solver='lbfgs',
    max_iter=100,
    multi_class='auto',
    verbose=0,
    warm_start=False,
    n_jobs=None,
    l1_ratio=None, random_state= 42), X_train, y_train, scoring='f1_weighted')
```

```python
f1_scores.mean()
```

```python
# Подберите значение константы регуляризации C в логистической регрессии, перебирая гиперпараметр от 0.001 до 100 включительно, проходя по степеням 10.
# Для выбора С примените перебор по сетке по тренировочной выборке (GridSearchCV из библиотеки sklearn.model_selection) с тремя фолдами и метрикой качества - f1_weighted.
# Остальные параметры оставьте по умолчанию. В ответ запишите наилучшее среди искомых значение С.
from sklearn.model_selection import GridSearchCV

parameters = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

clf = GridSearchCV(LogisticRegression(penalty='l2', dual=False, tol=0.0001, fit_intercept=True, intercept_scaling=1, class_weight=None,  solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None, random_state=42),
                   cv=3, scoring='f1_weighted', param_grid=parameters)
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
# Добавьте в тренировочные и тестовые данные новый признак ‘ram_resolution’, вычисляемый по формуле ram * (px_height + px_width)
# На тренировочных данных с новым признаком заново с помощью GridSearchCV (с тремя фолдами и метрикой качества - f1_weighted) подберите оптимальное значение С
# (перебирайте те же значения С, что и в предыдущих заданиях), в ответ напишите наилучшее качество алгоритма по метрике f1_weighted (можно посмотреть, обратившись к полю best_score_ обученного GridSearchCV),
# ответ округлите до сотых.
X_train['ram_resolution'] = X_train['ram'] * (X_train['px_height'] + X_train['px_width'])
X_test['ram_resolution'] = X_test['ram'] * (X_test['px_height'] + X_test['px_width'])
```

```python
clf = GridSearchCV(LogisticRegression(penalty='l2', dual=False, tol=0.0001, fit_intercept=True, intercept_scaling=1, class_weight=None,  solver='lbfgs', max_iter=100, multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None, random_state=42),
                   cv=3, scoring='f1_weighted', param_grid=parameters)
clf.fit(X_train, y_train)
```

```python
clf.best_score_
```
